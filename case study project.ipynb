```python
import pandas as pn
import matplotlib.pyplot as plt
import collections
df=pn.read_csv('startup_funding.csv')
df['CityLocation'].replace("bangalore", "Bangalore", inplace = True)
df['CityLocation'].replace("Delhi", "New Delhi", inplace = True)
df['CityLocation'].replace("SFO / Bangalor", "Bangalore", inplace = True)
df['CityLocation'].replace("Seattle / Bangalore", "Bangalore", inplace = True)
df=df['CityLocation'].dropna()
dic={}
for index, row in df.items():
    ls = row.split('/')
    location = ls[0].strip()
    lll=['Bangalore','New Delhi','Mumbai','Gurgaon', 'Noida']
    if location in lll:    
        if location in dic:
            dic[location] += 1
        else:
            dic[location] = 1
os=dict(sorted(dic.items(), key=lambda item: item[1], reverse=True)[:10])
plt.pie(os.values(),labels=os.keys(),autopct='%1.2f%%')
plt.show()
    
```

```python
import pandas as pn 
import matplotlib.pyplot as plt
import collections
df=pn.read_csv('startup_funding.csv')
df['InvestorsName'] = df['InvestorsName'].astype(str)
dic={}
l=[]
for index,rows in df.iterrows():
    l=rows['InvestorsName'].split(',')
    for i in l:
        i=i.strip();
        if i in dic:
            dic[i]+=1
        else :
            dic[i]=1
od=dict(sorted(dic.items(),key=lambda item:item[1],reverse=True))
ld=dict(list(od.items())[:5])

plt.pie(ld.values(),labels=ld.keys(),autopct='%1.2f%%')
plt.show()
plt.plot(ld.keys(),ld.values(),'b--')
plt.scatter(ld.keys(),ld.values())
plt.xticks(rotation = 40)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv')

df['InvestorsName'].replace("Undisclosed investors", "Undisclosed Investors", inplace=True)
df['InvestorsName'].replace("Undisclosed Investors", "", inplace=True)
df['StartupName'].replace('Oyo Rooms', 'Oyo', inplace=True)
df['StartupName'].replace("OyoRooms", "Oyo", inplace=True)
df['StartupName'].replace("Oyorooms", "Oyo", inplace=True)
df['StartupName'].replace("OYO Rooms", "Oyo", inplace=True)
df['StartupName'].replace('Paytm Marketplace', 'Paytm', inplace=True)
df['StartupName'].replace('Flipkart.com', 'Flipkart', inplace=True)
df['StartupName'].replace('Ola Cabs', 'Ola', inplace=True)

df.dropna(inplace=True, subset=["InvestorsName", "StartupName"])
df.reset_index(drop=True, inplace=True)

Investor_name = df['InvestorsName']
startup_name = df['StartupName']
cnt = {}
name_startup = {}
n = len(Investor_name)

for idx in range(n):
    i = Investor_name[idx].strip()
    name = startup_name[idx].strip()
    string = i.strip().split(',')
    for j in string:
        k = j.strip()
        if k != "":
            if k not in name_startup:
                cnt[k] = 1
                li = [name]
                name_startup[k] = li
            else:
                li = name_startup[k]
                if name not in li:
                    cnt[k] += 1
                    li.append(name)
                    name_startup[k] = li

od = dict(sorted(cnt.items(), key=lambda item: item[1], reverse=True))
ld = dict(list(od.items())[:5])

plt.pie(ld.values(), labels=ld.keys(), autopct='%1.2f%%')
plt.show()
plt.plot(ld.keys(), ld.values(), 'b--')
plt.scatter(ld.keys(), ld.values())
plt.xticks(rotation=40)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('startup_funding.csv')
df['InvestmentType'].replace('SeedFunding','Seed Funding',inplace=True)
df['InvestmentType'].replace('PrivateEquity','Private Equity',inplace=True)
df['InvestmentType'].replace('Crowd funding','Crowd Funding',inplace=True)
df.dropna(subset=['InvestmentType'],inplace=True)
df['InvestorsName'].replace("Undisclosed investors", "Undisclosed Investors", inplace=True)
df['InvestorsName'].replace("Undisclosed Investors", "", inplace=True)
df['StartupName'].replace('Oyo Rooms', 'Oyo', inplace=True)
df['StartupName'].replace("OyoRooms", "Oyo", inplace=True)
df['StartupName'].replace("Oyorooms", "Oyo", inplace=True)
df['StartupName'].replace("OYO Rooms", "Oyo", inplace=True)
df['StartupName'].replace('Paytm Marketplace', 'Paytm', inplace=True)
df['StartupName'].replace('Flipkart.com', 'Flipkart', inplace=True)
df['StartupName'].replace('Ola Cabs', 'Ola', inplace=True)
df.dropna(inplace=True, subset=["InvestorsName", "StartupName"])
df.reset_index(drop=True, inplace=True)
cc={}
ll={}
jj=df['InvestorsName']
ss=df['StartupName']
bb=df['InvestmentType']
n=len(jj)
for indx in range(n):
    if bb[indx]=='Seed Funding' or bb[indx]=='Crowdfunding' :
        name=ss[indx].strip()
        i=jj[indx].strip()
        sss=i.split(',')
        for j in sss:
            ff=j.strip()
            if ff != "":
                if ff not in ll:
                    cc[ff]=1
                    aa=[name]
                    ll[ff]=aa
                else :
                    lo=ll[ff]
                    if name not in lo:
                        cc[ff]+=1
                        lo.append(name)
                        ll[ff]==lo
od = dict(sorted(cc.items(), key=lambda item: item[1], reverse=True))
ld = dict(list(od.items())[:5])

plt.pie(ld.values(), labels=ld.keys(), autopct='%1.2f%%')
plt.show()
plt.bar(ld.keys(), ld.values())
plt.scatter(ld.keys(), ld.values())
plt.grid()
plt.xticks(rotation=40)
plt.show()
```
